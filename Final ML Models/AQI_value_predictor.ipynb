```python
dist=(poll['AQI Category'])
distset=set(dist)
dd=list(distset)
dictOfWords = {dd[i]: i for i in range(len(dd))}
poll['AQI Category']=poll['AQI Category'].map(dictOfWords)
```

## Data Splitting

```python
x = poll[['Country','City','CO AQI Value','Ozone AQI Value','NO2 AQI Value','PM2.5 AQI Value']]
y = poll[['AQI Value','AQI Category']]
```

```python
from sklearn.preprocessing import StandardScaler

x = poll[['Country', 'City', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']]

scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)

# Convert the scaled data back to a DataFrame for better readability
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x_scaled,y,test_size=0.2, random_state=2)
```

## Model:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn import tree
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression

regr=RandomForestRegressor(max_depth=10,random_state=0)
regr.fit(xtrain,ytrain)
print(regr.predict(xtest))
```

```python
y_pred = regr.predict(xtest)
y_pred[:,0].dtype
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ytest,y_pred)
```

```python
def check_user_input():
    try:
        user_input = float(input("Enter the AQI Value: "))  
        
        if 0< user_input<50:
            print("The condition of air is GOOD!!!! Congratulations you breathe healthy air!!!!")
        elif 50<user_input< 100:
            print("The condition of air is MODERATE!!!! A little care is needed try planting trees!!!! .")
        elif 100< user_input< 150:
            print("The air is unhealthy for some sensitive groups!!! @@ Plant more trees ...Check air pollution guidelines.")
            
        elif 150 < user_input <200:
            print("The air is unhralthy for all!!! On a serious note, take care of environment!!!Some members of the general public may experience health effects; members of sensitive groups may experience more serious health effects. @@ Plant more trees, Fit air filters in industries etc"
                 )
        elif 200< user_input<250:
            print("The air is very unhealthy!!! Health alert: The risk of health effects is increased for everyone.")
            
        else:
            print("The air is hazardous!!!..Health warning of emergency conditions: everyone is more likely to be affected.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")


check_user_input()
```
